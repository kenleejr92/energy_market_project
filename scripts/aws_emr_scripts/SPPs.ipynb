```python
# Import all libraries needed for the tutorial

# General syntax to import specific functions in a library: 
##from (library) import (specific library function)
from pandas import DataFrame, read_csv

# General syntax to import a library but no functions: 
##import (library) as (give the library a nickname/alias)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd #this is how I usually import pandas

df = pd.read_csv('/home/kenlee/Energy_Market_Project/LMP_data/cdr.00012331.0000000000000000.20160421.132355.DAMSPNP4190.csv')
del df['DSTFlag']

sorted = df.sort_values(['SettlementPoint'])

nodes = np.array_split(sorted, 610)  # split the dataframe into 24 hr chunks

nodes_SPPs = []
for node in nodes:
    SPPs = node[['DeliveryDate','SettlementPointPrice']].values
    column_list = ['DeliveryDate', node.iloc[0]['SettlementPoint']]
    new_node = pd.DataFrame(SPPs,index=node.HourEnding,columns=column_list)
    nodes_SPPs.append(new_node)

df_final = pd.concat(nodes_SPPs, axis=1,ignore_index=False)
df_final
```

```python
df2 = pd.read_csv('/home/kenlee/Energy_Market_Project/LMP_data/cdr.00012331.0000000000000000.20160420.123042.DAMSPNP4190.csv')
sorted2 = df2.sort_values(['SettlementPoint'])
nodes2 = np.array_split(sorted2, 610)  # split the dataframe into 24 hr chunks
nodes_SPPs2 = []
for node2 in nodes2:
    SPPs2 = node2[['DeliveryDate','SettlementPointPrice']].values
    column_list2 = ['DeliveryDate', node2.iloc[0]['SettlementPoint']]
    new_node2 = pd.DataFrame(SPPs2,index=node2.HourEnding,columns=column_list2)
    nodes_SPPs2.append(new_node2)
df2_final = pd.concat(nodes_SPPs2, axis=1,ignore_index=False)
```
